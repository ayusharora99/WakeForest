```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Batter_Profiles.csv')
```

```python
data = data.loc[:, data.columns != 'Unnamed: 0']
```

```python
data.head()
```

```python
unique_ids = data.BatterId.unique()
```

```python
len(unique_ids)
```

```python
df = pd.DataFrame(columns=['Batter_ID',
                           'Up_The_Middle_BIP',
                           'Up_The_Middle_Hit',
                           'Up_The_Middle_BIP+Hit',
                           'SS_BIP',
                           'SS_Hit',
                           'SS_BIP+Hit',
                           '3_4_Hole_BIP',
                           '3_4_Hole_Hit',
                           '3_4_Hole_BIP+Hit',
                           '1B_BIP',
                           '1B_Hit',
                           '1B_BIP+Hit',
                           '5_6_Hole_BIP',
                           '5_6_Hole_Hit',
                           '5_6_Hole_BIP+Hit',
                           '2B_BIP',
                           '2B_Hit',
                           '2B_BIP+Hit',
                           '3B_BIP',
                            '3B_Hit',
                          '3B_BIP+Hit'])
```

```python
for i in np.nditer(unique_ids):
    
    test_id = i
    test_batter = data[data.BatterId == i]
    
    cluster_0 = test_batter[test_batter['cluster #'] == 0]
    if len(cluster_0) > 0:
        cluster_0_prob = len(cluster_0) / len(test_batter)
        cluster_0_BIP = len(cluster_0)
        hits = cluster_0[cluster_0['PlayResult'] == 1]
        cluster_0_hit_prob = len(hits) / len(cluster_0)
        cluster_0_hits = len(hits)
        test_batter_hit_prob_cluster_0 = cluster_0_prob * cluster_0_hit_prob * 100
    
    cluster_1 = test_batter[test_batter['cluster #'] == 1]
    if len(cluster_1) > 0:
        cluster_1_prob = len(cluster_1) / len(test_batter)
        cluster_1_BIP = len(cluster_1)
        hits = cluster_1[cluster_1['PlayResult'] == 1]
        cluster_1_hit_prob = len(hits) / len(cluster_1)
        cluster_1_hits = len(hits)
        test_batter_hit_prob_cluster_1 = cluster_1_prob * cluster_1_hit_prob * 100
    
    cluster_2 = test_batter[test_batter['cluster #'] == 2]
    if len(cluster_2) > 0:
        cluster_2_prob = len(cluster_2) / len(test_batter)
        cluster_2_BIP = len(cluster_2)
        hits = cluster_2[cluster_2['PlayResult'] == 1]
        cluster_2_hit_prob = len(hits) / len(cluster_2)
        cluster_2_hits = len(hits)
        test_batter_hit_prob_cluster_2 = cluster_2_prob * cluster_2_hit_prob * 100
    
    cluster_3 = test_batter[test_batter['cluster #'] == 3]
    if len(cluster_3) > 0:
        cluster_3_prob = len(cluster_3) / len(test_batter)
        cluster_3_BIP = len(cluster_3)
        hits = cluster_3[cluster_3['PlayResult'] == 1]
        cluster_3_hit_prob = len(hits) / len(cluster_3)
        cluster_3_hits = len(hits)
        test_batter_hit_prob_cluster_3 = cluster_3_prob * cluster_3_hit_prob * 100
    
    cluster_4 = test_batter[test_batter['cluster #'] == 4]
    if len(cluster_4) > 0:
        cluster_4_prob = len(cluster_4) / len(test_batter)
        cluster_4_BIP = len(cluster_4)
        hits = cluster_4[cluster_4['PlayResult'] == 1]
        cluster_4_hit_prob = len(hits) / len(cluster_4)
        cluster_4_hits = len(hits)
        test_batter_hit_prob_cluster_4 = cluster_4_prob * cluster_4_hit_prob * 100
        
    
    cluster_5 = test_batter[test_batter['cluster #'] == 5]
    if len(cluster_5) > 0:
        cluster_5_prob = len(cluster_5) / len(test_batter)
        cluster_5_BIP = len(cluster_5)
        hits = cluster_5[cluster_5['PlayResult'] == 1]
        cluster_5_hit_prob = len(hits) / len(cluster_5)
        cluster_5_hits = len(hits)
        test_batter_hit_prob_cluster_5 = cluster_5_prob * cluster_5_hit_prob * 100
    
    cluster_6 = test_batter[test_batter['cluster #'] == 6]
    if len(cluster_6) > 0:
        cluster_6_prob = len(cluster_6) / len(test_batter)
        cluster_6_BIP = len(cluster_6)
        hits = cluster_6[cluster_6['PlayResult'] == 1]
        cluster_6_hit_prob = len(hits) / len(cluster_6)
        cluster_6_hits = len(hits)
        test_batter_hit_prob_cluster_6 = cluster_6_prob * cluster_6_hit_prob * 100
    
    df = df.append({'Batter_ID': test_id,
                    'SS_BIP': cluster_0_BIP,
                    'SS_Hit':cluster_0_hits,
                    'SS_BIP+Hit': test_batter_hit_prob_cluster_0,
                    'Up_The_Middle_BIP':cluster_1_BIP,
                    'Up_The_Middle_Hit':cluster_1_hits,
                    'Up_The_Middle_BIP+Hit':test_batter_hit_prob_cluster_1,
                    '2B_BIP':cluster_2_BIP,
                    '2B_Hit':cluster_2_hits,
                    '2B_BIP+Hit':test_batter_hit_prob_cluster_2,
                    '5_6_Hole_BIP':cluster_3_BIP,
                    '5_6_Hole_Hit':cluster_3_hits,
                    '5_6_Hole_BIP+Hit':test_batter_hit_prob_cluster_3,
                    '3_4_Hole_BIP':cluster_4_BIP,
                    '3_4_Hole_Hit':cluster_4_hits,
                    '3_4_Hole_BIP+Hit':test_batter_hit_prob_cluster_4,
                    '3B_BIP':cluster_5_BIP,
                    '3B_Hit':cluster_5_hits,
                    '3B_BIP+Hit':test_batter_hit_prob_cluster_5,
                    '1B_BIP':cluster_6_BIP,
                    '1B_Hit':cluster_6_hits,
                    '1B_BIP+Hit':test_batter_hit_prob_cluster_6}, ignore_index=True)
    test_id = 0
    cluster_0_prob = 0
    cluster_0_BIP = 0
    cluster_0_hits = 0
    cluster_0_hit_prob = 0
    test_batter_hit_prob_cluster_0 = 0
    
    cluster_1_prob= 0
    cluster_1_BIP = 0
    cluster_1_hits = 0
    cluster_1_hit_prob= 0
    test_batter_hit_prob_cluster_1 = 0
    
    cluster_2_prob= 0
    cluster_2_BIP = 0
    cluster_2_hits = 0
    cluster_2_hit_prob= 0
    test_batter_hit_prob_cluster_2 = 0
    
    cluster_3_prob= 0
    cluster_3_BIP = 0
    cluster_3_hits = 0
    cluster_3_hit_prob= 0
    test_batter_hit_prob_cluster_3 = 0
    
    cluster_4_prob= 0
    cluster_4_BIP = 0
    cluster_4_hits = 0
    cluster_4_hit_prob= 0
    test_batter_hit_prob_cluster_4 = 0
    
    cluster_5_prob= 0
    cluster_5_BIP = 0
    cluster_5_hits = 0
    cluster_5_hit_prob= 0
    test_batter_hit_prob_cluster_5 = 0
    
    cluster_6_prob= 0
    cluster_6_BIP = 0
    cluster_6_hits = 0
    cluster_6_hit_prob= 0
    test_batter_hit_prob_cluster_6 = 0
    
    hits = 0
```

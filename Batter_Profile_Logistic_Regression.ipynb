```python
# Calculate Expected Hits per hitter per Cluster with appropriate model on each
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Batter_Profiles.csv')
middle = pd.read_csv('middle_logistic_regression.csv')
ss = pd.read_csv('SS_logistic_regression.csv')
three_four = pd.read_csv('3_4_logistic_regression.csv')
first_base = pd.read_csv('1B_logistic_regression.csv')
five_six = pd.read_csv('5_6_logistic_regression.csv')
second_base = pd.read_csv('2B_logistic_regression.csv')
third_base = pd.read_csv('3B_logistic_regression.csv')
```

```python
unique_ids = data.BatterId.unique()
```

```python
df = pd.DataFrame(columns=['Batter_ID',
                           'BIP_middle',
                           'actual_hits_middle',
                           'expected_hits_middle',
                           'shifted_hits_middle',
                           'BIP_SS',
                           'actual_hits_SS',
                           'expected_hits_SS',
                           'shifted_hits_SS',
                           'BIP_3_4',
                           'actual_hits_3_4',
                           'expected_hits_3_4',
                           'shifted_hits_3_4'
                           'BIP_1B',
                           'actual_hits_1B',
                           'expected_hits_1B',
                           'shifted_hits_1B',
                           'BIP_5_6',
                           'actual_hits_5_6',
                           'expected_hits_5_6',
                           'shifted_hits_5_6'
                           'BIP_2B',
                           'actual_hits_2B',
                           'expected_hits_2B',
                           'shifted_hits_2B'
                           'BIP_3B',
                           'actual_hits_3B',
                           'expected_hits_3B',
                           'shifted_hits_3B'
                           'BIP',
                           'total_expected_hits',
                           'total_shifted_hits'])
```

```python
for i in np.nditer(unique_ids):
    
    test_id = i
    
    test_batter = middle[middle.BatterId == test_id]
    BIP_middle = len(test_batter)
    actual_hits_middle = sum(test_batter.PlayResult_y)
    expected_hits_middle = sum(test_batter.hit_prob)
    shifted_hits_middle = sum(test_batter.shifted_hit_prob)

    test_batter = ss[ss.BatterId == test_id]
    BIP_SS = len(test_batter)
    actual_hits_SS = sum(test_batter.PlayResult_y)
    expected_hits_SS = sum(test_batter.hit_prob)
    shifted_hits_SS = sum(test_batter.shifted_hit_prob)
    
    test_batter = three_four[three_four.BatterId == test_id]
    BIP_3_4 = len(test_batter)
    actual_hits_3_4 = sum(test_batter.PlayResult_y)
    expected_hits_3_4 = sum(test_batter.hit_prob)
    shifted_hits_3_4 = sum(test_batter.shifted_hit_prob)
    
    test_batter = first_base[first_base.BatterId == test_id]
    BIP_1B = len(test_batter)
    actual_hits_1B = sum(test_batter.PlayResult_y)
    expected_hits_1B = sum(test_batter.hit_prob)
    shifted_hits_1B = sum(test_batter.shifted_hit_prob)
    
    test_batter = five_six[five_six.BatterId == test_id]
    BIP_5_6 = len(test_batter)
    actual_hits_5_6 = sum(test_batter.PlayResult_y)
    expected_hits_5_6 = sum(test_batter.hit_prob)
    shifted_hits_5_6 = sum(test_batter.shifted_hit_prob)
    
    test_batter = second_base[second_base.BatterId == test_id]
    BIP_2B = len(test_batter)
    actual_hits_2B = sum(test_batter.PlayResult_y)
    expected_hits_2B = sum(test_batter.hit_prob)
    shifted_hits_2B = sum(test_batter.shifted_hit_prob)
    
    test_batter = third_base[third_base.BatterId == test_id]
    BIP_3B = len(test_batter)
    actual_hits_3B = sum(test_batter.PlayResult_y)
    expected_hits_3B = sum(test_batter.hit_prob)
    shifted_hits_3B = sum(test_batter.shifted_hit_prob)
    
    total_expected_hits = expected_hits_middle + expected_hits_SS + expected_hits_3_4 +expected_hits_1B+expected_hits_5_6+expected_hits_2B+expected_hits_3B 
    total_shifted_hits = shifted_hits_middle + shifted_hits_SS + shifted_hits_3_4 + shifted_hits_1B + shifted_hits_5_6 + shifted_hits_2B + shifted_hits_3B    
    BIP = BIP_middle + BIP_SS + BIP_3_4 + BIP_1B + BIP_5_6 + BIP_2B + BIP_3B 
    
    df = df.append({'Batter_ID':test_id,
                           'BIP_middle':BIP_middle ,
                           'actual_hits_middle': actual_hits_middle,
                           'expected_hits_middle': expected_hits_middle,
                           'shifted_hits_middle': shifted_hits_middle,
                           'BIP_SS':BIP_SS,
                           'actual_hits_SS': actual_hits_SS,
                           'expected_hits_SS': expected_hits_SS,
                           'shifted_hits_SS': shifted_hits_SS,
                           'BIP_3_4': BIP_3_4,
                           'actual_hits_3_4': actual_hits_3_4,
                           'expected_hits_3_4': expected_hits_3_4,
                           'shifted_hits_3_4': shifted_hits_3_4,
                           'BIP_1B': BIP_1B,
                           'actual_hits_1B': actual_hits_1B,
                           'expected_hits_1B': expected_hits_1B,
                           'shifted_hits_1B': shifted_hits_1B,
                           'BIP_5_6': BIP_5_6,
                           'actual_hits_5_6': actual_hits_5_6,
                           'expected_hits_5_6': expected_hits_5_6,
                           'shifted_hits_5_6': shifted_hits_5_6,
                           'BIP_2B': BIP_2B,
                           'actual_hits_2B':actual_hits_2B,
                           'expected_hits_2B': expected_hits_2B,
                           'shifted_hits_2B': shifted_hits_2B,
                           'BIP_3B': BIP_3B,
                           'actual_hits_3B': actual_hits_3B,
                           'expected_hits_3B': expected_hits_3B,
                           'shifted_hits_3B':shifted_hits_3B,
                           'BIP': BIP,
                           'total_expected_hits': total_expected_hits,
                           'total_shifted_hits': total_shifted_hits}, ignore_index=True)
    BIP_middle = 0
    actual_hits_middle = 0
    expected_hits_middle = 0
    shifted_hits_middle = 0
    BIP_SS = 0
    actual_hits_SS = 0
    expected_hits_SS = 0
    shifted_hits_SS = 0
    BIP_3_4 = 0
    actual_hits_3_4 = 0
    expected_hits_3_4 = 0
    shifted_hits_3_4 = 0
    BIP_1B = 0
    actual_hits_1B = 0
    expected_hits_1B = 0
    shifted_hits_1B = 0
    BIP_5_6 = 0
    actual_hits_5_6 = 0
    expected_hits_5_6 = 0
    shifted_hits_5_6 = 0
    BIP_2B = 0
    actual_hits_2B = 0
    expected_hits_2B = 0
    shifted_hits_2B = 0
    BIP_3B = 0
    actual_hits_3B = 0
    expected_hits_3B = 0
    shifted_hits_3B = 0
    BIP = 0
    total_expected_hits = 0
    total_shifted_hits = 0
```

```python
df.to_csv('Batters_Profile_Logistic_Regression.csv')
```
